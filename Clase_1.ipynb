```python
import os
import pandas as pd

# Lista de directorios donde se encuentran los archivos CSV
directories = [
    "/home/mdiaz/workshop_carneiro/RawData/Demengeot",
    "/home/mdiaz/workshop_carneiro/RawData/Howard",
    "/home/mdiaz/workshop_carneiro/RawData/HowardNew",
    "/home/mdiaz/workshop_carneiro/RawData/Vilanova"
]

# Crear una lista para almacenar los datos
data = []

# Iterar sobre cada directorio
for directory in directories:
    # Iterar sobre todos los archivos en el directorio actual
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            # Leer el archivo CSV
            file_path = os.path.join(directory, filename)
            df = pd.read_csv(file_path)

            # Obtener la información del nombre del archivo
            title = filename
            date = filename.split('_')[0]  # Parte antes de "_plate"
            lab = os.path.basename(directory)  # Nombre del directorio
            plate = filename.split('plate')[1].split('.csv')[0].strip()  # Obtener la parte que empieza con "plate" hasta antes de ".csv"
  # Obtener el número del plate

            # Calcular la media y desviación estándar de las columnas 1 y 2
            mean_col1 = df.iloc[:, 0].mean()
            mean_col2 = df.iloc[:, 1].mean()
            sd_col1 = df.iloc[:, 0].std()
            sd_col2 = df.iloc[:, 1].std()

            # Añadir los datos a la lista
            data.append({
                "Title": title,
                "Date": date,
                "Lab": lab,
                "Plate": f"plate_{plate}",  # Formato "plate_{número}"
                "Mean_Col1": mean_col1,
                "Mean_Col2": mean_col2,
                "S.D._Col1": sd_col1,
                "S.D._Col2": sd_col2
            })

# Crear un DataFrame con todos los datos
df_summary = pd.DataFrame(data)

# Guardar el DataFrame en un archivo CSV si es necesario
df_summary.to_csv('/home/mdiaz/workshop_carneiro/ToolingSystemsBiology/summary.csv', index=False)

print(df_summary)
```

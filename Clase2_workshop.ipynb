```python
import os
import pandas as pd

# Lista de directorios donde se encuentran los archivos CSV
directories = [
    "/home/mdiaz/workshop_carneiro/RawData/Demengeot",
    "/home/mdiaz/workshop_carneiro/RawData/Howard",
    "/home/mdiaz/workshop_carneiro/RawData/HowardNew",
    "/home/mdiaz/workshop_carneiro/RawData/Vilanova"
]

# Crear una lista para almacenar los datos
data = []

# Iterar sobre cada directorio
for directory in directories:
    # Iterar sobre todos los archivos en el directorio actual
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            # Leer el archivo CSV
            file_path = os.path.join(directory, filename)
            df = pd.read_csv(file_path)

            # Obtener la informacin del nombre del archivo
            title = filename
            date = filename.split('_')[0]  # Parte antes de "_plate"
            lab = os.path.basename(directory)  # Nombre del directorio
            plate = filename.split('plate')[1].split('.csv')[0].strip()  # Obtener el nmero del plate

            # Calcular la media y desviacin estndar de las columnas 1 y 2
            mean_col1 = df.iloc[:, 0].mean()
            mean_col2 = df.iloc[:, 1].mean()
            sd_col1 = df.iloc[:, 0].std()
            sd_col2 = df.iloc[:, 1].std()

            # Aadir los datos a la lista
            data.append({
                "Title": title,
                "Date": date,
                "Lab": lab,
                "Plate": f"plate_{plate}",  # Formato "plate_{nmero}"
                "Mean_Col1": mean_col1,
                "Mean_Col2": mean_col2,
                "S.D._Col1": sd_col1,
                "S.D._Col2": sd_col2
            })

# Crear un DataFrame con todos los datos
df_summary = pd.DataFrame(data)

# Guardar el DataFrame en un archivo CSV si es necesario
df_summary.to_csv('/home/mdiaz/workshop_carneiro/ToolingSystemsBiology/summary.csv', index=False)

print(df_summary)
```

```python
import pandas as pd
import os

# Inicializa la lista para almacenar la informacin
data = []

# Directorios de los archivos
directories = ['Demengeot', 'Howard', 'HowardNew', 'Vilanova']

# Recorre cada directorio
for lab in directories:
    folder_path = f'/home/mdiaz/workshop_carneiro/RawData/{lab}'
    
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            
            # Extrae el Title, Date y Lab
            title = filename
            date = filename.split('_plate')[0]  # Obtiene la parte antes de "_plate"
            plate = filename.split('plate')[1].split('.csv')[0].strip()  # Obtiene "plate_x" antes de ".csv"

            # Calcula la media y desviacin estndar de las columnas 1 y 2 (ltima fila, excluyendo la ltima columna)
            last_row = df.iloc[-1, :-1]  # ltima fila, excluyendo la ltima columna
            
            mean_last_row = last_row.mean()  # Media de la ltima fila
            std_last_row = last_row.std()  # Desviacin estndar de la ltima fila
            
            # Almacena los datos en la lista
            data.append([title, date, lab, plate, mean_last_row, std_last_row])

# Crea el DataFrame con los datos
df_summary_background = pd.DataFrame(data, columns=['Title', 'Date', 'Lab', 'Plate', 'Mean_Last_Row', 'S.D._Last_Row'])

# Muestra el DataFrame resultante
print(df_summary_background)

```

```python
import os
import pandas as pd

# Directorio base donde estn las subcarpetas
base_dir = '/home/mdiaz/workshop_carneiro/RawData'

# Lista para almacenar las rutas completas de los archivos CSV
csv_files = []

# Buscar recursivamente en las subcarpetas
for root, _, files in os.walk(base_dir):
    for file in files:
        if file.endswith('.csv'):
            csv_files.append(os.path.join(root, file))

# Asumiendo que merged_df ya est definido y tiene las columnas adecuadas
# Ejemplo de cmo podra verse merged_df
# merged_df = pd.DataFrame({'Lab': ['Demengeot', 'Howard'], 'Title': ['file1.csv', 'file2.csv'], 'Threshold': [1.0, 2.0]})

# Diccionario que asocia cada laboratorio con su carpeta
lab_paths = {
    "Demengeot": "/home/mdiaz/workshop_carneiro/RawData/Demengeot/",
    "Howard": "/home/mdiaz/workshop_carneiro/RawData/Howard/",
    "HowardNew": "/home/mdiaz/workshop_carneiro/RawData/HowardNew/",
    "Vilanova": "/home/mdiaz/workshop_carneiro/RawData/Vilanova/"
}

# Crear una columna con la ruta completa del archivo CSV
merged_df['csv_path'] = merged_df.apply(lambda row: lab_paths[row['Lab']] + row['Title'], axis=1)

# Lista de columnas que deseas extraer
columns_to_extract = ["Trit_Ref1", "Trit_Ref2", "Trit_Ctrl1", "Trit_Ctrl2", "Trit_Infec1", "Trit_Infec2", "Trit_Infec3", "Trit_Infec4"]

# Inicializar las nuevas columnas en merged_df
for col in columns_to_extract:
    merged_df[col] = None

# Iterar sobre cada archivo y su correspondiente Threshold
for idx, csv_path in enumerate(merged_df['csv_path']):
    try:
        # Cargar el archivo csv
        data = pd.read_csv(csv_path)
        
        # Obtener el valor de Threshold correspondiente en merged_df
        threshold_value = merged_df.loc[idx, 'Threshold']
        
        # Crear una lista para almacenar los valores extrados de cada columna
        extracted_values = []
        
        # Iterar sobre las primeras 8 columnas de datos
        for col in data.columns[:8]:
            # Calcular la diferencia entre Threshold y cada valor en la columna
            diff = threshold_value - data[col]
            
            # Buscar la primera fila desde abajo donde la diferencia es positiva
            positive_diff = diff[::-1][diff > 0]  # Filtrar las diferencias positivas en orden inverso
            
            if not positive_diff.empty:  # Comprobar que hay valores positivos
                row_index = positive_diff.index[0]  # Obtener el primer ndice
                # Extraer el valor original de esta fila y columna
                extracted_value = data.loc[row_index, col]
                extracted_values.append(extracted_value)
            else:
                extracted_values.append(None)  # Agregar None si no hay diferencias positivas
            
            # Debugging output
            print(f"Archivo: {csv_path}, Columna: {col}, Threshold: {threshold_value}, "
                  f"Diferencias: {diff.tolist()}, Positivas: {positive_diff.tolist()}")
        
        # Asignar los valores extrados a las columnas correspondientes en merged_df
        merged_df.loc[idx, columns_to_extract] = extracted_values

    except FileNotFoundError as e:
        print(f"Archivo no encontrado: {e}")

# Verificar el resultado
print(merged_df[columns_to_extract].head())
```
